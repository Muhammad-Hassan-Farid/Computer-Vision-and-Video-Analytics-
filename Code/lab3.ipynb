```python
import cv2

import matplotlib.pyplot as plt

# Image path 
img_path = 'Images/OIP (2).webp'

# Image Size
img_size = (512,512)

# Read in GrayScle
image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

resize_image = cv2.resize(image, img_size)

plt.imshow(resize_image, cmap='gray')

plt.axis('off')

plt.show()
```

```python
sobel_x = cv2.Sobel(resize_image, cv2.CV_64F, 1, 0, ksize=3)

sobel_y = cv2.Sobel(resize_image, cv2.CV_64F, 0, 1, ksize=3)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)

plt.imshow(sobel_x, cmap='gray')

plt.title('Sobel X')

plt.axis('off')

plt.subplot(1,2,2)

plt.imshow(sobel_y, cmap='gray')

plt.title('Sobel Y')

plt.axis('off')

plt.show()
```

```python
import numpy as np

gradient_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)

gradient_magnitude = np.uint8(gradient_magnitude)

plt.imshow(gradient_magnitude, cmap='gray')

plt.title('Gradient Magnitude')

plt.axis('off')

```

```python
laplacian = cv2.Laplacian(resize_image, cv2.CV_64F)

plt.imshow(laplacian, cmap='gray')

plt.title('Laplacian Edge Detection')

plt.axis('off')

plt.show()
```
